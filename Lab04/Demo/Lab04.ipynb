```r
demoDataf<- as.data.frame(read_excel("../../Dataset/demoDataset.xlsx"))
```

What are the variance and the standard deviation of the daily mean temperature
of HKO?

### Answer:

```r
var(demoDataf$hko_mean_temp )
sd(demoDataf$hko_mean_temp )
```

    variance = 27.059

### Question 5

What are the upper and the lower quartiles of the daily mean temperature of
HKO?

### Answer:

```r
quantile (demoDataf$hko_mean_temp,c(0.25,0.75),type=6)
```

    
    Upper quartile = 27.875
    Lower quartile = 18.825


### Question 6

What are the skewness and the kurtosis of the daily mean temperature of HKO?

### Answer:

```r
library(e1071)
skewness(demoDataf$hko_mean_temp,type=2)
kurtosis(demoDataf$hko_mean_temp,type=2)
```

    
    Skewness = -0.196
    Kurtosis = -1.169


### Question 7

Plot a line graph for the monthly mean of the daily mean temperature of HKO.

### Answer:

```r
ag <- aggregate(hko_mean_temp ~ month, data=demoDataf, mean)
plot(ag$month, ag$hko_mean_temp, type="l",  
     xlab="Month", ylab="Mean daily mean temperature of HKO", 
     main="Line Chart of Mean Temperature vs Month")
```

### Question 8

Plot a boxplot (with four boxes) of the wind speed of HKO for diﬀerent cardinal
directions.

### Answer:

```r
boxplot(hko_wind_speed ~ hko_wind_cardinal_dir,data=demoDataf,
        main="Wind speed of different cardinal directions", 
        xlab="Direction", ylab="Wind speed", names=c("N","E","S","W"))
```
